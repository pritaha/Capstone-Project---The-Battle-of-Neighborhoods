```python
xl = pd.ExcelFile("madrid_restbars.xlsx")
xl.sheet_names
[u'Sheet1']
madrid_restbars = xl.parse("Sheet1")
madrid_restbars.head()
```

```python
madrid_drinking = madrid_restbars['Venue Category'].isin(['Beer bar', 'Bar', 'Gastropub', 'Sports Bar', 
                                                        'Wine Bar', 'Cocktail Bar', 'Beer Garden', 
                                                        'Brewery', 'Pub', 'Irish Pub', 'Karaoke Bar'
                                                       ])
```

Remove all venues that are not places specifically to drink or serve alcohol

```python
madrid_drinking = madrid_restbars[madrid_drinking]
```

```python
madrid_drinking.head()
```

```python
writer = ExcelWriter('madrid_drinking.xlsx')
madrid_drinking.to_excel(writer,'Sheet1')
writer.save()
```
